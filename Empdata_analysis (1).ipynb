Performing Scatter plot

Moving left to right → employees with less to more experience

Moving bottom to top → employees with lower to higher income

Every point shows the income of an employee at a specific experience level



 The points are widely scattered

There is no strong linear relationship

Monthly income varies significantly at every experience level








```python
sns.scatterplot(x="Years_of_Experience", y="Monthly_Income",data=df)
plt.show()
```

```python

sns.scatterplot(x=df['Monthly_Income'],y=df['Years_of_Experience'],hue=df['Gender'],style=df['Department'],size=df['Performance_Rating'])

plt.xlabel("Monthly Income")
plt.ylabel("Years of Experience")
plt.title("Monthly Income vs Experience by Gender, Department, and Performance")
plt.show()
```

Bar plot

```python
sns.barplot(x=df['Department'],y=df['Monthly_Income'])

plt.xlabel("Department")
plt.ylabel("Average Monthly Income")
plt.title("Average Monthly Income by Department")
plt.show()
```
